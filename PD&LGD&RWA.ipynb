```python
lgd_x1 = [[-1, 0.5982, 1.075872],
              [0.5982, 0.6261, 0.342808],
              [0.6261, 0.7130, 0.014483],
              [0.7130, 0.7747, -0.407573],
              [0.7747, 10, -0.934577]]

lgd_x2 =  [[-1,170020.5334,-1.740796],
           [170020.5334,356913.6863,-0.296552],
           [356913.6863,429858.6742,0.022997],
           [429858.6742,673170.3249,0.265285],
           [673170.3249,99999999,0.880985]]

lgd_x3 =  [[0,630,-0.570725],
           [630,710,-0.173790],
           [710,939,0.151167],
           [939,999,0.567175]] 

lgd_x4 =  [[-1,0.0622,0.077273],
           [0.0622,0.0699,-0.169878],
           [0.0699,1,-0.592054]]

x5 = [['BC', 0.401136], 
          ['YT', 0.401136], 
          ['QC', 0.255954],
          ['ON', 0.182497],
          ['SK', 0.002673],
          ['MB', 0.002673],
          ['NU', 0.002673],
          ['NT', 0.002673],
          ['AB', -0.480898],
          ['NB', -1.263872],
          ['NS', -1.263872],
          ['PE', -1.263872],
          ['NL', -1.263872]]

lgd_x5 = pd.DataFrame(x5, columns= ['Province', 'LGD_Province_WOE'])

```

```python
for i in lgd_x1:
    df.loc[(df.LTV_Incl_Part_Prov_WF > i[0])&(df.LTV_Incl_Part_Prov_WF <= i[1]), 'LGD_LTV_WOE'] = i[2]
df['LGD_LTV_WOE'] = df['LGD_LTV_WOE'].fillna(-0.934577)

for i in lgd_x2:
    df.loc[(df.Appr_Val_WF > i[0])&(df.Appr_Val_WF <= i[1]), 'LGD_Appr_Val_WOE'] = i[2]
df['LGD_Appr_Val_WOE'] = df['LGD_Appr_Val_WOE'].fillna(-1.740796)

for i in lgd_x3:
    df.loc[(df.Max_BNI_Score_App > i[0])&(df.Max_BNI_Score_App <= i[1]), 'LGD_BNI_WOE'] = i[2]
df['LGD_BNI_WOE'] = df['LGD_BNI_WOE'].fillna(-0.570725)
    
for i in lgd_x4:
    df.loc[(df.Interest_Rate > i[0])&(df.Interest_Rate <= i[1]), 'LGD_Interest_Rate_WOE'] = i[2]

df2 = df.merge(lgd_x5, on = 'Province', how = 'left')
df2['LGD_UR_WOE'] = df2['Unemployment_diff_2Q']
lgd_coeff = [-0.816026,-0.932599,-0.813700,-0.737364,-0.653026,-0.777661,0.187656]
```

```python
lgd_input = df2.loc[:, df2.columns.str.startswith('LGD')]
lgd_input.insert(0, 'Intercept', 1)
Log_Odds_LGD = np.dot(lgd_input, np.array(lgd_coeff).transpose()).astype(float)
LGD = 1 / (1 + np.exp(-1 * Log_Odds_LGD))
df2['LGD_Final']= LGD
```

```python
def RWA_cal(PD,LGD,EAD, Default_Ind):
    corre = 0.15
    if Default_Ind == 1:
        capital_require = 0.8
    else:
        capital_require = LGD*st.norm.cdf((1-corre)**(-0.5) * st.norm.ppf(PD) + (corre / (1-corre)) ** (0.5) * st.norm.ppf(0.999)) - PD*LGD
    RWA_Final = capital_require * 12.5 * EAD
    
    return RWA_Final
```

```python
RWA_Final = df2.apply(lambda s: RWA_cal(s['PD_Final']\
                                        ,s['LGD_Final']\
                                        ,s['RemainingPrincipal_excl_Partner']\
                                        ,s['Default_Ind']), axis=1)
RWA_Final
```

```python
df2['RWA_Final'] = RWA_Final
```
